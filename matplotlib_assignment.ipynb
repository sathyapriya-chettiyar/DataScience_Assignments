```python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(y,x,marker='o',linestyle='-',color='r')
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("Line plot")
plt.grid(True)
plt.show()

#quest 2
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students,marks,color=['red','orange','blue','green'])
plt.xlabel("students")
plt.ylabel("Marks")
plt.title("Marks scored by students")
plt.show()

#question 3
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1, 0, 0, 0]
plt.pie(revenue,labels=regions,autopct="%1.1f%%" ,explode=explode)

plt.title("companyâ€™s revenue")
plt.show()

#qest 4
sample=np.random.randint(101,size=1000)
plt.hist(sample,bins=10,color="blue",edgecolor='black')
plt.xlabel("integers (1-100)")
plt.ylabel("frequency")
plt.title("Random Integer")
plt.show()

```
