4. Reshape a 1D array of size 12 into a 3 X 4 2D array and slice the first two rows and last two columns.  


```python
import numpy as np
#quetion 1
oneDArr=np.arange(1,20)
np.random.shuffle(oneDArr)
print("one D Array",oneDArr)
print("Sum of the Array : ",np.sum(oneDArr))
print("Mean of the Array : ",np.mean(oneDArr))
print("Median of the Array : ",np.median(oneDArr))
print("Standard deviation of the Array : ",np.std(oneDArr))
indice=np.where(oneDArr >10)[0]
print("element > 10",indice)

#quetion 2
twoDArr=np.arange(16).reshape(4,4)
print("two D array\n",twoDArr)
print("Transpose\n",twoDArr.T)

print("Row Wise Sum",np.sum(twoDArr,axis=1))
print("Cloumn Wise Sum",np.sum(twoDArr,axis=0))

#ques 3
array1=np.random.randint(1,20,(3,3))
array2=np.random.randint(1,20,(3,3))
print("array1\n",array1)
print("array2\n",array2)
print("Additon\n",array1+array2)
print("Subtraction\n",array1-array2)
print("Multiplication\n",array1*array2)
print("Dot Product\n",np.dot(array1,array2))

#quest 4
arr=np.arange(12)
arr_reshape=arr.reshape(3,4)
print("reshape 1D to 2D\n",arr_reshape)
arr_reshape[0:2, 2:]
```
